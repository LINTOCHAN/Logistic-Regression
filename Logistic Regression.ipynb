```python
#imputation with mode for the column employment type
data['Employment.Type']=data['Employment.Type'].fillna(data['Employment.Type'].mode()[0])
```

```python
num_dt=data.select_dtypes(exclude=['object'])              #getting numerical values
```

```python
num_dt.head()
```

```python
cat_dt=data.select_dtypes(include=['object'])             #getting categorical values
```

```python
cat_dt.head()
```

```python
#importing labelencoder
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
#label encoding except date of birth and disbursal date
for i in cat_dt:
    if i!='Date.of.Birth' and i!='DisbursalDate':
        cat_dt[i]=le.fit_transform(cat_dt[i])
```

```python
cat_dt.head()
```

```python
dt1=pd.concat([cat_dt,num_dt],axis=1)   #concatenating both numerical and categorical
```

```python
dt1.head()
```

```python
# import the date class 
from datetime import date
import datetime 
#defining function called age
def age(birth):
    birth=datetime.strptime(birth,'%d-%m-%y')
    today=date.today()
    age=today.year-birth.year-((today.month,today.day)<(birth.month,birth.day))
    return(age)
```

```python
#importing datetime
from datetime import datetime
#converting date of birth into age and dropping date of birth
dob=dt1['Date.of.Birth']
dt1['Age']=dob.map(age)
dt1.drop(['Date.of.Birth'],axis=1,inplace=True)
```

```python
dt1.drop(['DisbursalDate'],axis=1,inplace=True)            #dropping disbursal date
```

```python
dt1.head()
```

```python
target=dt1['loan_default']               #assigning target variable
```

```python
dt1.drop('loan_default',axis=1,inplace=True)           #dropping target variable
```

```python
#train test splitting the dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(dt1,target, test_size=0.3,random_state=42)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

reg = LogisticRegression() 
  
# train the model using the training sets 
reg.fit(x_train, y_train) 
print(reg.score(x_train, y_train))            #accuracy of the training
```

```python
y_pred=reg.predict(x_test)              #predicting
```

```python
print(reg.score(x_test, y_test))               #accuracy for test data
```
